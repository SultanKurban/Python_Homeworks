```python
import pandas as pd 

df = pd.read_csv('tackoverflow_qa.csv')
df.head()

```

```python
# 1. Find all questions that were created before 2014

df["creationdate"] = pd.to_datetime(df["creationdate"])

# Filter all rows from year 2014
result = df[df["creationdate"].dt.year < 2014]
result
print(result)
```

```python
# 2. Find all questions with a score more than 50
my_filter_2 = df.score > 50
df[my_filter_2]
print(df[my_filter_2])

```

```python
# 3. Find all questions with a score between 50 and 100
my_filter_3 = df.score.between(50,100)
df[my_filter_3]
print(df[my_filter_3])
```

```python
# 4. Find all questions answered by Scott Boston

my_filter_4 = df.ans_name == 'Scott Boston'

df[my_filter_4]

df[df.ans_name == 'Scott Boston']
```

```python
# 5. Find all questions answered by the following 5 users

users_5 = ['Adam Hughes', 'joris', 'unutbu', 'Avaris', 'Scott Boston']

my_filter_5 = df[df['ans_name'].isin(users_5)]

my_filter_5
```

```python
# 6. Find all questions that were created between March, 2014 and October 2014 that were answered by Unutbu and have score less than 5.
df["creationdate"] = pd.to_datetime(df["creationdate"])

data_filter = (df['creationdate'] >= '2014-03-01') & (df['creationdate'] <= '2014-10-31')
answer_filter = df['ans_name'] == 'Unutbu'
score_filter = df['score'].lt(5)

filter_all6 = data_filter & answer_filter & score_filter

df[filter_all6]
```

```python
# 7. Find all questions that have score between 5 and 10 or have a view count of greater than 10,000

my_filter_score = df.score.between(5,10)

my_filter_view_count = df.viewcount > 10000

all_filters = my_filter_score | my_filter_view_count

df[all_filters]
```

```python
# 8. Find all questions that are not answered by Scott Boston
my_filter_8 = df.ans_name != 'Scott Boston'
df[my_filter_8]

df[df.ans_name != "Scott Boston"]
```

```python
titanic_df = pd.read_csv("titanic.csv")
titanic_df.head()
```

```python
# 1. Select Female Passengers in Class 1 with Ages between 20 and 30: Extract a DataFrame containing female passengers in Class 1 with ages between 20 and 30.

filter_class = titanic_df['Pclass'] == 1 
filter_age = titanic_df['Age'].between(20, 30)
filter_sex = titanic_df['Sex'] == 'female'

all_filters_11 = filter_class & filter_age & filter_sex
titanic_df[all_filters_11]
```

```python
# 2. Filter Passengers Who Paid More than $100: Create a DataFrame with passengers who paid a fare greater than $100.

filter_paid = titanic_df['Fare'].gt(100)

titanic_df[filter_paid]
```

```python
# 3. Select Passengers Who Survived and Were Alone: Filter passengers who survived and were traveling alone (no siblings, spouses, parents, or children).

titanic_df.head()

filter_survived = titanic_df['Survived'] == 1
filter_alone = titanic_df['SibSp'] == 0
filter_parents = titanic_df['Parch'] == 0

all_filters_3 = filter_survived & filter_alone & filter_parents

titanic_df[all_filters_3]
```

```python
# 4. Filter Passengers Embarked from 'C' and Paid More Than $50: Create a DataFrame with passengers who embarked from 'C' and paid more than $50.

titanic_df.head()

filter_embarked = titanic_df['Embarked'] == 'C'
filter_paidd = titanic_df['Fare'] > 50

all_filter_4 = filter_embarked & filter_paidd

titanic_df[all_filter_4]
```

```python
# 5. Select Passengers with Siblings or Spouses and Parents or Children: Extract passengers who had both siblings or spouses aboard and parents or children aboard.

titanic_df.head()

filter_SibSp = titanic_df['SibSp'] == 1
filter_Parch = titanic_df['Parch'] == 1

all_filter_5 = filter_SibSp & filter_Parch

titanic_df[all_filter_5]
```

```python
# 6. Filter Passengers Aged 15 or Younger Who Didn't Survive: Create a DataFrame with passengers aged 15 or younger who did not survive.
titanic_df.head()

age_filter = titanic_df['Age'].le(15)
survive_filter = titanic_df['Survived'] == 0

all_filter_6 = age_filter & survive_filter
titanic_df[all_filter_6]
```

```python
# 7. Select Passengers with Cabins and Fare Greater Than $200: Extract passengers with known cabin numbers and a fare greater than $200.
titanic_df.head()

Cabin_filter = titanic_df['Cabin'].ne('NaN')
Fare_filter = titanic_df['Fare'] > 200

all_filter_7 = Cabin_filter & Fare_filter
titanic_df[all_filter_7]
```

```python
# 8. Filter Passengers with Odd-Numbered Passenger IDs: Create a DataFrame with passengers whose PassengerId is an odd number.
titanic_df.head()

odd_id = titanic_df['PassengerId'] % 2 != 0

titanic_df[odd_id]
```
