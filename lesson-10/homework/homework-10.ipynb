```python
"""Homework 1. ToDo List Application

Define Task Class:
Create a Task class with attributes such as task title, description, due date, and status.
Define ToDoList Class:
Create a ToDoList class that manages a list of tasks.
Include methods to add a task, mark a task as complete, list all tasks, and display incomplete tasks.
Create Main Program:
Develop a simple CLI to interact with the ToDoList.
Include options to add tasks, mark tasks as complete, list all tasks, and display only incomplete tasks.
Test the Application:
Create instances of tasks and test the functionality of your ToDoList."""

class Task:

    def __init__(self, title, description, due_date, status):
        self.title = title
        self.description = description
        self.due_date = due_date
        self.status = status
    
class ToDoList:
    def __init__(self):
        self.task = []

    def add_task(self, task):
        self.task.append(task)

    def mark_complete(self, task):
        self.task[self.task.index[task]].status = 'completed'
    
    def list_all_tasks(self):
        for task in self.tasks:
            print(f'{task.title} is {task.status} and its due date is {task.due_date}')
    
    def incomplete_tasks(self):
        for task in self.tasks:
            if task.status.lower() == 'incomplete':
                print(f'{task.title.capitalize()} is {task.status.capitalize()} and its due date is {task.due_date}')

to_do_app = ToDoList()


def main():
    while True:
        user_input = int(input('\n What do you want to do with this ToDoList App \n 1. Add a new task \n 2. Mark a task as complete \n 3. List all tasks \n 4. List incomplete tasks \n 5. Exit \n Choose your option'))

        if user_input == 1:
            task_title = input('Enter the title of task \n')
            task_description = input('Enter the description of the task \n')
            task_due_date = input('Enter the due date of the task \n')
            task_status = input('Enter the status of the task \n')

            user_task = Task(task_title, task_description, task_due_date, task_status)

            to_do_app.add_task(user_task)
            print('Your task is added succesfully')
        
        elif user_input == 2:
            print('Your task complete')

        elif user_input == 3:
            to_do_app.list_all_tasks()
        
        elif user_input == 4:
            to_do_app.incomplete_tasks()
        
        elif user_input == 5:
            print('You are leaving the app')
            break
        else:
            print('You entered an invalid input')
            user_input = int(input('\n What do you want to do with this ToDoList App \n 1. Add a new task \n 2. Mark a task as complete \n 3. List all tasks \n 4. List incomplete tasks \n 5. Exit \n Choose your option'))
main()
```
