```python
#Homework Assignment 1: Analyzing Sales Data

#Tasks:

# 1.Group the data by the Category column and calculate the following aggregate statistics for each category:
        #Total quantity sold.
        #Average price per unit.
        #Maximum quantity sold in a single transaction.
import pandas as pd
sales_data = pd.read_csv('sales_data.csv')

print(sales_data.groupby('Category')['Quantity'].sum())
print(sales_data.groupby('Product')['Quantity'].sum())
print(sales_data.groupby('Date')['Quantity'].max())


# 2.Identify the top-selling product in each category based on the total quantity sold.
import pandas as pd
sales_data = pd.read_csv('sales_data.csv')

top_sell = sales_data.loc[sales_data.groupby('Category')['Quantity'].idxmax()]
print(top_sell)

# 3.Find the date on which the highest total sales (quantity * price) occurred.
import pandas as pd
sales_data['Total'] = sales_data['Quantity'] * sales_data['Price']

highest = sales_data.loc[sales_data.groupby('Date')['Total'].idxmax()]

print(highest)
```

```python
#Homework Assignment 2: Examining Customer Orders

#Tasks:
import pandas as pd

# 1.Group the data by CustomerID and filter out customers who have made less than 20 orders.
customer_orders = pd.read_csv('customer_orders.csv')

print(customer_orders.groupby('CustomerID')['OrderID'].count().reset_index().query('OrderID < 20'))

# 2.Identify customers who have ordered products with an average price per unit greater than $120.
customer_orders = pd.read_csv('customer_orders.csv')

print(customer_orders.groupby('CustomerID')['Price'].mean().reset_index().query('Price > 120'))

# 3.Find the total quantity and total price for each product ordered, and filter out products that have a total quantity less than 5 units.

customer_orders = pd.read_csv('customer_orders.csv')

print(customer_orders.groupby('Product')['Quantity'].sum())
print(customer_orders.groupby('Product')['Price'].sum())
print(customer_orders.groupby('Product')['Quantity'].sum().reset_index().query('Quantity < 5'))

```
